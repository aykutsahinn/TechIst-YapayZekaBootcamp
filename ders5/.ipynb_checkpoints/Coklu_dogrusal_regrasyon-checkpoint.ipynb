```python
#Regresyon ve veri önişleme kütüpahneleri yükle

#veri setini eğitim ve test olarak bölmek için kütüphane
from sklearn.model_selection import train_test_split

#Ölçeklendirme kütüphanesi içe aktar
from sklearn.preprocessing import StandardScaler as Scaler

#lineer regresyon modelini içe aktar
from sklearn.linear_model import LinearRegression

#Modeli değerlendirmek için metrikler
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Özellikler ve hedef sütunlarını belirleyeceğiz
X=df[['TV','Radio','Newspaper']] #X=df.drop('Sales',axis=1)
y=df['Sales']
```

```python
#Veri setini eğitim ve test olarak ayır
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               test_size=.2,#test verisinin büyüklüğü %20
                                               random_state=42) 
# random_state=42 Rastgele olarak belirle ama 42  kullanan herkese aynı sonuçları ver
```

```python
#Model oluşturma
lr=LinearRegression() #bir linear regresyon nesnesi oluştur
```

```python
#modeli fit et eğit
lr.fit(X_train,y_train) #eğitim setinin X ve y sini ver
```

```python
#Tahminleri yap
y_predict=lr.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
#modelin tahminleri ve gerçek değerlerin karşılaştırılması
model_predict_actual=pd.DataFrame({'Gerçek':y_test,'Tahmin':y_predict})
model_predict_actual
```

```python
#Hatayı da sütun oalrak ekle
model_predict_actual['Error']=abs(model_predict_actual['Gerçek']-model_predict_actual['Tahmin'])
model_predict_actual
```

```python
#modeli değerlendir
print ("R2 Score : ",r2_score(y_test,y_predict)) 
#r2_score(y_test,y_predict) gerçek değerlerle tahminleri karşılaştır
```

```python
#RMSE Hata Kareler Ortalaması
print("RMSE : ",mean_squared_error(y_test,y_predict)**.5)
```

Bu tip çalışmalarda ölçkelnedirme yapmak önemlidir.
Bazen sonucu etkilemez. 
Ama ölçeklendirme farklı aralıklardaki özelliklerin modelimizi domine etmesini engeller.
Örnek: 
            TV Radyo Newspaper 
(147.0425, 23.264000000000006, 30.553999999999995)

Hangi Ölçeklendirme Yöntemi Ne Zaman Kullanılır?
Min-Max Ölçeklendirme (Normalizasyon):
Değerleri 0 ile 1 arasına sıkıştırır.

Kullanılması önerilen durumlar:
Değişken ölçekleri birbirinden çok farklı olduğunda.
Algoritmanın performansını artırmak istendiğinde.
Z-Skor Ölçeklendirme (Standardizasyon):
Ortalamayı 0, standart sapmayı 1 yapar.

Kullanılması önerilen durumlar:
Değişkenlerin göreceli değerlerinin önemli olduğu durumlarda.
Aykırı değerlerin etkisini azaltmak istendiğinde.

Log Dönüşümü:
Değerleri logaritmik olarak dönüştürür.
Kullanılması önerilen durumlar:
Değerlerde çok büyük farklılıklar olduğunda.
Veride asimetri (çarpıklık) olduğunda.

```python
#Ölçeklendirme Standard Scaler
sc=Scaler()
X_scaler=sc.fit_transform(X)
X_scaler
```

```python
#Eğitim ve test verileri
X_train,X_test,y_train,y_test=train_test_split(X_scaler,y,
                                               test_size=.2,#test verisinin büyüklüğü %20
                                               random_state=42) 
```

```python
#Model oluştur
mlr=LinearRegression()
```

```python
#modeli eğitiyoruz
mlr.fit(X_train,y_train)
```

```python
#tahminleri alıyoruz
y_pred=mlr.predict(X_test)
```

```python
#Modelin R2 değeri
r2_score(y_test,y_pred)
```

```python
#MSE ve RMSE hata metrikleri
mean_squared_error(y_test,y_pred)**.5
```

```python
import pickle
#modeli kaydet
pickle.dump(mlr,open('mlr_model.pkl','wb'))
```
